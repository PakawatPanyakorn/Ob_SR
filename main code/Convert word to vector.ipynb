```python
def find_replace_multi(string, dictionary):
    for item in dictionary.keys():
        # sub item for item's paired value in string
        string = re.sub(item, dictionary[item], string)
    return string

def word2list(word):
    word = word.lower().strip()
    dicts = {',' : '',
             '-' : ' ',
             "'" : '',
             '\(' : '',
             '\)' : '',
             ' and' : '',
             ' of ' : ' ',
             ' to ' : ' ',
             'centre' : 'center',
             'counselling' : 'counseling',
             'draughtspersons' : 'checkers player',
             'labourers' : 'laborer',
             'sericulturists' : 'sericulture',
             'quarriers' : 'quarry',
             'shotfirers' : 'blaster',
             'varnishers' : 'varnish',
             'fumigators' : 'fumigate',
             'coremakers' : 'core makers',
             'flamecutters' : 'flame cutters',
             'hammersmiths' : 'hammer smiths',
             'fellmongers' : 'fell mongers',
             'shoemaking' : 'shoe making',
             'labelling' : 'label',
             'apiarists' : 'beekeeper',
             'moulders' : 'moulder'
            }
    try: 
        return find_replace_multi(word, dicts).split(' ').remove('')
    except:
        return find_replace_multi(word, dicts).split(' ')
```
